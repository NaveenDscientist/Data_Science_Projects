## What problem are you trying to answer?

The problem is to **predict the level of damage (damage grade)** that a building is likely to suffer after an earthquake based on its **structural characteristics, materials used, geographical location, and usage information**.

Specifically, using historical data of buildings affected by earthquakes, the model aims to classify each building into one of three damage categories:

* **Low damage (Grade 1)**
* **Medium damage (Grade 2)**
* **Severe damage (Grade 3)**

This is formulated as a **multiclass classification problem**, where the goal is to learn patterns between building features and observed earthquake damage.

---

## What does success look like?

Success is achieved when the model can **accurately and consistently predict the damage grade of buildings** on unseen data.

From a practical and technical perspective, success means:

* The model achieves **high predictive performance** (good accuracy and F1-score) across all three damage classes.
* The model generalizes well, avoiding overfitting to historical earthquake data.
* Predictions are **reliable enough to support decision-making** in disaster preparedness and response.
* Important building features influencing damage (e.g., age, floors, construction material) are clearly identified and interpretable.
* The final solution is **efficient and scalable**, making it suitable for real-world deployment in resource-constrained environments.

●	· 	geo_level_1_id, geo_level_2_id, geo_level_3_id (type: int): geographic region in which building exists, from largest (level 1) to most specific sub-region (level 3). Possible values: level 1: 0-30, level 2: 0-1427, level 3: 0-12567.<br>

## 1. Introduction

This Exploratory Data Analysis (EDA) was conducted as part of the **Earthquake Damage Prediction** project. The primary objective of the EDA is to understand the structure, quality, and patterns within the dataset, identify relationships between features, and prepare the data for effective modeling. The target variable represents the **level of damage caused by an earthquake**, while the features describe building characteristics, location-related attributes, and structural conditions.

---

## 2. Dataset Overview

### 2.1 Shape and Structure

* The dataset consists of a **large number of observations**, each corresponding to an individual building.
* Features are a mix of **numerical and categorical variables**.
* The target variable is **categorical and multi-class**, representing different damage grades.

### 2.2 Feature Categories

The variables can be broadly grouped into:

* **Building characteristics**: age of the building, height, number of floors, foundation type, roof type, and construction material.
* **Geographical/location indicators**: region or administrative identifiers.
* **Structural and safety features**: presence of superstructure components, secondary usage indicators, and building condition flags.
* **Target variable**: damage grade (severity level).

---

## 3. Data Quality Assessment

### 3.1 Missing Values

* Most features contain **no or very minimal missing values**, indicating good data collection quality.
* A few categorical columns show sparse missingness, which can be handled using:

  * Mode imputation for categorical variables
  * Domain-informed defaults where applicable

### 3.2 Duplicate Records

* No significant number of duplicate rows were observed.
* Each record uniquely represents a building, making the dataset suitable for supervised learning.

### 3.3 Data Types

* Some categorical variables were initially encoded as integers.
* These variables were treated as **categorical**, not ordinal, during analysis and modeling.

---

## 4. Univariate Analysis

### 4.1 Target Variable (Damage Grade)

* The target variable shows **class imbalance**.
* One damage grade (moderate damage) appears most frequently.
* Severe and low damage classes are comparatively underrepresented.

**Insight:**
Class imbalance indicates the need for:

* Stratified sampling
* Appropriate evaluation metrics (F1-score, weighted metrics)
* Possible use of class weights or resampling techniques

### 4.2 Numerical Features

Key observations:

* **Building age**: Right-skewed distribution; older buildings are more prone to severe damage.
* **Number of floors / height**: Lower-rise buildings are more common, but damage severity increases with height in certain cases.
* **Continuous structural measures**: Presence of outliers, especially in age and height-related variables.

**Handling decisions:**

* Outliers were retained since they carry real-world significance (old or tall buildings).
* Scaling was applied where required for model compatibility.

### 4.3 Categorical Features

* Construction material and foundation type show **uneven frequency distributions**.
* Traditional or weaker materials appear more frequently in highly damaged buildings.

---

## 5. Bivariate Analysis

### 5.1 Target vs Numerical Features

* **Building age vs damage**: Clear trend showing higher damage grades for older buildings.
* **Number of floors vs damage**: Multi-story buildings tend to show higher damage probabilities.

### 5.2 Target vs Categorical Features

* **Foundation type**: Certain foundation types are strongly associated with lower damage grades.
* **Roof type and superstructure indicators**: Reinforced structures correlate with reduced damage severity.

**Insight:**
Structural quality plays a more important role than location alone in predicting damage severity.

---

## 6. Multivariate Analysis

### 6.1 Correlation Analysis (Numerical Features)

* Moderate correlation observed among building height, number of floors, and age.
* No extreme multicollinearity detected, making tree-based models suitable.

### 6.2 Feature Interactions

* Older buildings with weak construction materials show the **highest probability of severe damage**.
* Reinforced structures mitigate damage even for taller buildings.

---

## 7. Class Imbalance Analysis

* The imbalance in the target variable can bias models toward the majority class.
* This justified the use of:

  * `class_weight='balanced'`
  * Weighted evaluation metrics

## 8. Feature Engineering Insights from EDA

Based on EDA, the following feature engineering steps were motivated:

* Encoding categorical variables using **label encoding or one-hot encoding** depending on the model.
* Creating interaction-aware models (tree-based) rather than linear-only approaches.
