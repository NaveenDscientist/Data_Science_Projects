## What problem are you trying to answer?

The problem is to **predict the level of damage (damage grade)** that a building is likely to suffer after an earthquake based on its **structural characteristics, materials used, geographical location, and usage information**.

Specifically, using historical data of buildings affected by earthquakes, the model aims to classify each building into one of three damage categories:

* **Low damage (Grade 1)**
* **Medium damage (Grade 2)**
* **Severe damage (Grade 3)**

This is formulated as a **multiclass classification problem**, where the goal is to learn patterns between building features and observed earthquake damage.

---

## What does success look like?

Success is achieved when the model can **accurately and consistently predict the damage grade of buildings** on unseen data.

From a practical and technical perspective, success means:

* The model achieves **high predictive performance** (good accuracy and F1-score) across all three damage classes.
* The model generalizes well, avoiding overfitting to historical earthquake data.
* Predictions are **reliable enough to support decision-making** in disaster preparedness and response.
* Important building features influencing damage (e.g., age, floors, construction material) are clearly identified and interpretable.
* The final solution is **efficient and scalable**, making it suitable for real-world deployment in resource-constrained environments.

# Handling outliers

```python
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df[col] = np.clip(df[col], lower, upper)
```

**Insight**

Caps extreme values instead of removing rows

Preserves data volume

Prevents skewed model learning

# Separate Features and Target

```python
X = df.drop(columns=['damage_grade'])
y = df['damage_grade']
```

# Age-Based Risk Features

# Building age categories

```python
X['age_group'] = pd.cut(
    X['age'],
    bins=[0, 10, 30, 60, 1000],
    labels=[0, 1, 2, 3]
)
```

**Insight**

Older buildings follow outdated safety standards

Grouping reduces noise and improves generalization

# Old building risk flag

```python
X['is_old_building'] = (X['age'] > 30).astype(int)
```

**Insight**

Simple binary risk indicator

Helps linear models and trees alike

# Height & Floor Vulnerability Features

# Floor-to-height ratio

```python
X['floor_height_ratio'] = (
    X['count_floors_pre_eq'] / (X['height_percentage'] + 1)
)
```

**Insight**

Tall buildings with many floors are more unstable

Ratio captures vertical stress better than raw values

# High-rise indicator

```python
X['is_high_rise'] = (X['count_floors_pre_eq'] >= 4).astype(int)
```

**Insight**

High-rise buildings behave differently during earthquakes

# Area Utilization Features

# Density indicator

```python
X['area_height_interaction'] = (
    X['area_percentage'] * X['height_percentage']
)
```

**Insight**

Dense + tall structures experience higher seismic load

# Overcrowding proxy

```python
X['people_density_proxy'] = (
    X['count_families'] / (X['area_percentage'] + 1)
)
```

**Insight**

More families in smaller areas often indicates weaker construction

# Superstructure Material Risk Scores

# Weak material count

```python
weak_materials = [
    'has_superstructure_adobe_mud',
    'has_superstructure_mud_mortar_stone',
    'has_superstructure_mud_mortar_brick'
]

X['weak_material_count'] = X[weak_materials].sum(axis=1)
```

**Insight**

Mud/adobe materials perform poorly during earthquakes

# Strong material count

```python
strong_materials = [
    'has_superstructure_rc_engineered',
    'has_superstructure_cement_mortar_brick',
    'has_superstructure_cement_mortar_stone'
]

X['strong_material_count'] = X[strong_materials].sum(axis=1)
```

**Insight**

Engineered RCC and cement structures reduce damage risk

# Material strength score

```python
X['material_strength_score'] = (
    X['strong_material_count'] - X['weak_material_count']
)
```

**Insight**

Single powerful composite feature

Positive → safer buildings

Negative → high-risk buildings

# Structural Complexity Features

# Mixed material indicator

```python
X['has_mixed_materials'] = (
    (X['weak_material_count'] > 0) &
    (X['strong_material_count'] > 0)
).astype(int)
```

**Insight**

Mixed construction leads to uneven load distribution

# Structural complexity score

```python
X['structural_complexity'] = (
    X['count_floors_pre_eq'] +
    X['weak_material_count'] +
    X['has_mixed_materials']
)
```

**Insight**

Higher complexity → higher seismic vulnerability

# Secondary Usage Risk Features

# Any secondary use flag

```python
X['has_any_secondary_use'] = (
    X[[col for col in X.columns if col.startswith('has_secondary_use_')]]
    .sum(axis=1) > 0
).astype(int)
```

**Insight**

Multi-purpose buildings are often not structurally optimized

# Public use risk flag

```python
public_use_cols = [
    'has_secondary_use_school',
    'has_secondary_use_health_post',
    'has_secondary_use_gov_office'
]

X['is_public_building'] = (
    X[public_use_cols].sum(axis=1) > 0
).astype(int)
```

**Insight**

Public buildings often house more people → higher risk impact
