## What problem are you trying to answer?

The problem is to **predict the level of damage (damage grade)** that a building is likely to suffer after an earthquake based on its **structural characteristics, materials used, geographical location, and usage information**.

Specifically, using historical data of buildings affected by earthquakes, the model aims to classify each building into one of three damage categories:

* **Low damage (Grade 1)**
* **Medium damage (Grade 2)**
* **Severe damage (Grade 3)**

This is formulated as a **multiclass classification problem**, where the goal is to learn patterns between building features and observed earthquake damage.

---

## What does success look like?

Success is achieved when the model can **accurately and consistently predict the damage grade of buildings** on unseen data.

From a practical and technical perspective, success means:

* The model achieves **high predictive performance** (good accuracy and F1-score) across all three damage classes.
* The model generalizes well, avoiding overfitting to historical earthquake data.
* Predictions are **reliable enough to support decision-making** in disaster preparedness and response.
* Important building features influencing damage (e.g., age, floors, construction material) are clearly identified and interpretable.
* The final solution is **efficient and scalable**, making it suitable for real-world deployment in resource-constrained environments.

# Data Cleaning

# Standardizing Column Names

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df.duplicated().sum()
```

# Missing Value Analysis

```python
missing_df = pd.DataFrame({
    "missing_count": df.isnull().sum(),
    "missing_percentage": df.isnull().mean() * 100
}).sort_values("missing_percentage", ascending=False)

missing_df
```

# Separating Numeric & Categorical Columns

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns
```

# Handling Missing Values

# Numerical → Median (robust to outliers)

```python
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

# Categorical → Mode

```python
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

# Outlier Treatment (IQR Method)

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.clip(df[col], lower, upper)
```

# Handling Incorrect / Invalid Values

```python
# Replacing negative values where logically impossible
for col in num_cols:
    df[col] = df[col].apply(lambda x: x if x >= 0 else np.nan)

# Refilling after correction
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

# Encode Categorical Variables

```python
# Label Encoding (Tree-based models friendly)

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

# Removing Highly Correlated Features

```python
corr_matrix = df.corr().abs()
upper_triangle = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

high_corr_cols = [
    col for col in upper_triangle.columns
    if any(upper_triangle[col] > 0.9)
]

df.drop(columns=high_corr_cols, inplace=True)
```

# Feature Engineering

# Identifying Feature Groups

```python
target = "damage_grade"

X = df.drop(columns=[target])
y = df[target]
```

# Building Age Features

Older buildings are more vulnerable.

```python
# Replacing zero construction years 
df["age"] = df["age"].replace(0, df["age"].median())

# Age buckets
df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 20, 40, 60, 1000],
    labels=[0, 1, 2, 3]
)
```

# Height & Floor Interaction Features

Tall buildings + weak materials = high damage.

```python
# Height per floor
df["height_per_floor"] = df["height_percentage"] / (df["count_floors_pre_eq"] + 1)

# Floors squared (non-linear effect)
df["floors_squared"] = df["count_floors_pre_eq"] ** 2
```

# Structural Strength Indicators

```python
# Wall + roof interaction
df["wall_roof_interaction"] = (
    df["has_superstructure_mud_mortar_stone"] +
    df["has_superstructure_cement_mortar_stone"] +
    df["has_superstructure_timber"] +
    df["has_superstructure_bamboo"]
)

# Weak structure flag
df["weak_structure"] = (
    (df["has_superstructure_mud_mortar_stone"] == 1) &
    (df["has_superstructure_cement_mortar_brick"] == 0)
).astype(int)
```

# Geometry & Shape Features

Irregular buildings collapse more easily.
