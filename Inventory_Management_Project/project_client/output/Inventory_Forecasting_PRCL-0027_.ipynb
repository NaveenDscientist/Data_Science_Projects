# Creating customised target for prediction

## Sorting Data Properly

````python
df = df.sort_values(['vehicle_no', 'invoice_date'])
````

## Creating Next Visit Date

````python
df['next_invoice_date'] = df.groupby('vehicle_no')['invoice_date'].shift(-1)
````

This gets the next service date for each vehicle.

## Calculating Days Until Next Visit

````python
df['days_until_next_visit'] = (
    df['next_invoice_date'] - df['invoice_date']
).dt.days
````

## Creating JIT Target (30-Day Rule)

````python
df['target'] = (
    df['days_until_next_visit'] <= 30
).astype(int)
````

## Removing Rows Where Next Visit Is Unknown

````python
# Last visit per vehicle has no future data
df = df.dropna(subset=['days_until_next_visit'])
````

# Verify Target Distribution

````python
df['target'].value_counts()
````

````python
X = df.drop(columns=[
    'target',
    'next_invoice_date',
    'days_until_next_visit'
])

y = df['target']
````

# What This Target Means For JIT

# If model predicts:

**Prediction	Meaning**<br>
1	Vehicle likely to return within 30 days → Stock parts<br>
0	Vehicle unlikely to return soon → Don't stock excess

**This helps service centre**:

*Maintain minimal inventory

*Reduce storage cost

*Prevent stockouts

# Making Multi-Class Target (Better JIT Strategy)

````python
def demand_class(days):
    if days <= 15:
        return 2  # High urgency
    elif days <= 30:
        return 1  # Medium urgency
    else:
        return 0  # Low urgency

df['target'] = df['days_until_next_visit'].apply(demand_class)
````

2 = Immediate stocking required

1 = Moderate

0 = Low

This is more powerful for inventory planning.

# Importing Necessary Libraries 

````python
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
````

# Stratified K-Fold (Important for Inventory Class Imbalance)

````python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
````

# Defining Optimized Models + Small Param Grid (Fast Execution)

## Logistic Regression

````python
log_model = LogisticRegression(max_iter=500, class_weight='balanced')

log_param = {
    'C': [0.1, 1],
    'solver': ['liblinear']
}
````

## Random Forest

````python
rf_model = RandomForestClassifier(random_state=42, n_jobs=-1)

rf_param = {
    'n_estimators': [100],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5]
}
````

## Extra Trees

````python
et_model = ExtraTreesClassifier(random_state=42, n_jobs=-1)

et_param = {
    'n_estimators': [100],
    'max_depth': [10, 20]
}
````

## Gradient Boosting

````python
gb_model = GradientBoostingClassifier(random_state=42)

gb_param = {
    'n_estimators': [100],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3]
}
````

# Create Model Dictionary

````python
models = {
    "Logistic Regression": (log_model, log_param),
    "Random Forest": (rf_model, rf_param),
    "Extra Trees": (et_model, et_param),
    "Gradient Boosting": (gb_model, gb_param)
}
````

# Run GridSearch + KFold (Optimized for Speed)

````python
best_models = {}

for name, (model, param) in models.items():
    print(f"\nTraining {name}...")

    grid = GridSearchCV(
        estimator=model,
        param_grid=param,
        cv=skf,
        scoring='f1',         
        verbose=0
    )

    grid.fit(X_train_smote, y_train_smote)  

    best_models[name] = grid
    print(f"Best Params for {name}: {grid.best_params_}")
````

# Final Test Evaluation 

````python
for name, model in best_models.items():
    
    print(f"\n{name} Test Performance")
    
    trained_features = model.best_estimator_.feature_names_in_
    
    # Align test set to training features
    X_test_aligned = X_test[trained_features]
    
    y_pred = model.best_estimator_.predict(X_test_aligned)
    
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred, zero_division=0))
````

# Collective accuracy of all models

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

model_scores = {}

for name, model in best_models.items():
    
    # Align features properly
    trained_features = model.best_estimator_.feature_names_in_
    X_test_aligned = X_test[trained_features]
    
    y_pred = model.best_estimator_.predict(X_test_aligned)
    
    acc = accuracy_score(y_test, y_pred)
    model_scores[name] = acc

# Convert to DataFrame
results_df = pd.DataFrame(model_scores.items(), columns=["Model", "Accuracy"])

# Sort by best
results_df = results_df.sort_values(by="Accuracy", ascending=False)

results_df
````

# Bar Plot Visualization

````python
plt.figure(figsize=(8,5))

plt.bar(results_df["Model"], results_df["Accuracy"])

plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Model Comparison for Inventory JIT Prediction")

plt.xticks(rotation=45)
plt.ylim(0,1)

plt.show()
````

# Multi-Metric Comparison Plot

````python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

metrics_data = []

for name, model in best_models.items():
    
    # Aligning feature names
    trained_features = model.best_estimator_.feature_names_in_
    X_test_aligned = X_test[trained_features]
    
    y_pred = model.best_estimator_.predict(X_test_aligned)
    
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    
    metrics_data.append([name, acc, precision, recall, f1])

# Creating DataFrame
metrics_df = pd.DataFrame(
    metrics_data,
    columns=["Model", "Accuracy", "Precision", "Recall", "F1-Score"]
)

metrics_df = metrics_df.sort_values(by="F1-Score", ascending=False)

metrics_df
````

````python
from sklearn.metrics import accuracy_score, classification_report

for name, model in best_models.items():
    
    trained_features = model.best_estimator_.feature_names_in_
    X_test_aligned = X_test[trained_features]
    
    y_pred = model.best_estimator_.predict(X_test_aligned)
    
    print(f"\n{name}")
    print("Best Params:", model.best_params_)
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred, zero_division=0))
````

````python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

metrics_list = []

for name, model in best_models.items():
    
    # Align feature names correctly (important)
    trained_features = model.best_estimator_.feature_names_in_
    X_test_aligned = X_test[trained_features]
    
    y_pred = model.best_estimator_.predict(X_test_aligned)
    
    # Some models may not support predict_proba
    if hasattr(model.best_estimator_, "predict_proba"):
        y_prob = model.best_estimator_.predict_proba(X_test_aligned)[:, 1]
        roc_auc = roc_auc_score(y_test, y_prob)
    else:
        roc_auc = np.nan
    
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    
    metrics_list.append([name, acc, precision, recall, f1, roc_auc])

# Creating DataFrame
metrics_df = pd.DataFrame(
    metrics_list,
    columns=["Model", "Accuracy", "Precision", "Recall", "F1-Score", "ROC-AUC"]
)

# Sorting by F1 Score (Best for JIT balance)
metrics_df = metrics_df.sort_values(by="F1-Score", ascending=False)

metrics_df
````

# Challenges Faced Report

### Challenge: Missing Values in Multiple Columns

**Problem:**
Columns like vehicle number, vehicle model, and invoice description contained missing values.

**Technique Used:**

* Used `fillna()` for categorical columns
* Used median imputation for numerical columns

**Reason:**
Median is robust to outliers, and categorical placeholders maintain dataset consistency.

---

### Challenge: Date Columns Causing Errors (NaT values)

**Problem:**
Date columns had invalid or inconsistent formats.

**Technique Used:**

```python
pd.to_datetime(errors='coerce')
```

**Reason:**
Converted invalid dates to NaT and handled them safely without breaking the model.

---

### Challenge: Standard Deviation Showing NaN for Dates

**Problem:**
`.describe()` showed NaN for date columns.

**Technique Used:**
Excluded datetime columns when calculating statistics.

**Reason:**
Standard deviation does not apply to datetime format.

---

### Challenge: Creating Target Variable

**Problem:**
No predefined classification target in dataset.

**Technique Used:**
Created target based on service frequency / time difference logic.

**Reason:**
JIT forecasting requires converting raw service data into actionable prediction labels.

---

### Challenge: Class Imbalance

**Problem:**
Target variable was imbalanced (more 0s than 1s).

**Technique Used:**
Applied **SMOTE (Synthetic Minority Oversampling Technique)**.

**Reason:**
Improves Recall and prevents model bias toward majority class.

---

### Challenge: Feature Engineering Complexity

**Problem:**
Raw columns were not predictive enough.

**Technique Used:**
Created features like:

* Service delay days
* Days since last service
* KM bucket categories
* Service type indicators

**Reason:**
Feature engineering improves model’s ability to capture service behavior patterns.

---

### Challenge: Feature Name Mismatch Error

**Problem:**
Model gave error about unseen feature names during prediction.

**Technique Used:**
Aligned test features using:

```python
trained_features = model.best_estimator_.feature_names_in_
X_test = X_test[trained_features]
```

**Reason:**
Ensures consistency between training and testing features.

---

### Challenge: Undefined Precision Warning

**Problem:**
Model did not predict some classes.

**Technique Used:**
Used:

```python
zero_division=0
```

**Reason:**
Prevents runtime warning and ensures stable metric output.

---

### Challenge: Slow Computation During GridSearch

**Problem:**
Full parameter grid caused slow performance.

**Technique Used:**
Reduced parameter search space and used 5-fold Stratified CV.

**Reason:**
Optimized performance for slow computing environment.

---

### Challenge: Selecting Best Model for JIT

**Problem:**
Accuracy alone was misleading.

**Technique Used:**
Used multi-metric comparison:

* Accuracy
* Precision
* Recall
* F1-score
* ROC-AUC

**Reason:**
For JIT systems, Recall and F1-score are more important than Accuracy.
