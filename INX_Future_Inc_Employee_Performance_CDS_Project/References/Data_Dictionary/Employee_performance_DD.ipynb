# Employee Performance Dataset - Data Dictionary

| S.No | Column Name              | Description                                                                 |
| ---- | ------------------------ | --------------------------------------------------------------------------- |
| 1    | Age                      | Age of the employee                                                         |
| 2    | Attrition                | Whether the employee left the company (Yes/No)                              |
| 3    | BusinessTravel           | Frequency of business travel (Travel_Rarely, Travel_Frequently, Non-Travel) |
| 4    | Department               | Department where the employee works                                         |
| 5    | DistanceFromHome         | Distance between employee’s home and workplace (in km)                      |
| 6    | Education                | Education level of the employee (numeric scale)                             |
| 7    | EducationField           | Field of study of the employee                                              |
| 8    | EmployeeNumber           | Unique identifier for each employee                                         |
| 9    | EnvironmentSatisfaction  | Satisfaction level with work environment (1–4 scale)                        |
| 10   | Gender                   | Gender of the employee                                                      |
| 11   | JobInvolvement           | Level of involvement in the job (1–4 scale)                                 |
| 12   | JobLevel                 | Job level or seniority of the employee                                      |
| 13   | JobRole                  | Role/designation of the employee                                            |
| 14   | JobSatisfaction          | Job satisfaction level (1–4 scale)                                          |
| 15   | MaritalStatus            | Marital status of the employee                                              |
| 16   | MonthlyIncome            | Monthly salary of the employee                                              |
| 17   | NumCompaniesWorked       | Number of companies the employee has worked for                             |
| 18   | OverTime                 | Whether the employee works overtime (Yes/No)                                |
| 19   | PercentSalaryHike        | Percentage increase in salary                                               |
| 20   | PerformanceRating        | Performance rating of the employee (2 = Low, 3 = Average, 4 = High)         |
| 21   | RelationshipSatisfaction | Satisfaction level with workplace relationships (1–4 scale)                 |
| 22   | TotalWorkingYears        | Total years of professional work experience                                 |
| 23   | TrainingTimesLastYear    | Number of training sessions attended in the last year                       |
| 24   | WorkLifeBalance          | Work-life balance rating (1–4 scale)                                        |
| 25   | YearsAtCompany           | Total years spent at the current company                                    |
| 26   | YearsInCurrentRole       | Years spent in the current job role                                         |
| 27   | YearsSinceLastPromotion  | Years since the last promotion                                              |
| 28   | YearsWithCurrManager     | Years working with the current manager                                      |
