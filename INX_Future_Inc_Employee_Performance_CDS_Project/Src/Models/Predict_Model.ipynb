## Importing Libraries

```python
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt  
```

## Loading Test Data

```python
X_test = pd.read_excel("X_test.xlsx")
y_test = pd.read_excel("y_test.xlsx")
```

## Loading Models

```python
lr_model = joblib.load("logistic_model.pkl")
dt_model = joblib.load("decision_tree_model.pkl")
rf_model = joblib.load("random_forest_model.pkl")
gb_model = joblib.load("gradient_boosting_model.pkl")
```

## Making Predictions

```python
lr_pred = lr_model.predict(X_test)
dt_pred = dt_model.predict(X_test)
rf_pred = rf_model.predict(X_test)
gb_pred = gb_model.predict(X_test)
```

## SAVING PREDICTIONS

```python
predictions = pd.DataFrame({
    'Actual': y_test.values.flatten(),
    'Logistic': lr_pred,
    'DecisionTree': dt_pred,
    'RandomForest': rf_pred,
    'GradientBoosting': gb_pred
})

predictions.to_excel("model_predictions_xtest_data.xlsx", index=False)
```

## Function to Plot & Save

```python
def plot_and_save_cm(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    
    plt.title(f"{model_name} Confusion Matrix")
    
    plt.savefig(f"{model_name}_cm.png")  
    plt.close()
```

## Generating & Save All

```python
plot_and_save_cm(y_test, lr_pred, "Logistic")
plot_and_save_cm(y_test, dt_pred, "DecisionTree")
plot_and_save_cm(y_test, rf_pred, "RandomForest")
plot_and_save_cm(y_test, gb_pred, "GradientBoosting")
```
