## Load Cleaned Dataset

```python
df = pd.read_excel("C:\\Users\\navee\\Downloads\\INX_Future_Inc_Employee_Performance_CDS_Project\\Data\\Processed\\employee_performance_processed.xlsx")

df.head()
```

# Handling Categorical Variables

## Identifying Categorical Columns

```python
cat_cols = df.select_dtypes(include='object').columns
print(cat_cols)
```

## Label Encoding (Binary Columns)

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

binary_cols = ['Attrition', 'Gender', 'OverTime']

for col in binary_cols:
    df[col] = le.fit_transform(df[col])
```

Converts Yes/No or Male/Female → 0/1<br>
Required for machine learning models

## One-Hot Encoding (Multi-category Columns)

```python
df = pd.get_dummies(df, columns=[
    'BusinessTravelFrequency',
    'EmpDepartment',
    'EducationBackground',
    'EmpJobRole',
    'MaritalStatus'
], drop_first=True)
```

Converts categorical variables into numeric columns<br>
drop_first=True avoids multicollinearity

# Feature Creation

## Experience Ratio

```python
df['ExperienceRatio'] = df['ExperienceYearsAtThisCompany'] / (df['TotalWorkExperienceInYears'] + 1)
```

Shows how long employee stayed in current company vs total experience<br>
Helps identify loyalty and stability

## Total Satisfaction Score

```python
df['TotalSatisfaction'] = (
    df['EmpJobSatisfaction'] +
    df['EmpEnvironmentSatisfaction'] +
    df['EmpRelationshipSatisfaction'] +
    df['EmpWorkLifeBalance']
)
```

Combines all satisfaction metrics into one<br>
Strong predictor of performance

## Average Tenure per Role

```python
df['AvgYearsPerRole'] = df['ExperienceYearsAtThisCompany'] / (df['ExperienceYearsInCurrentRole'] + 1)
```

Measures role stability<br>
Helps identify job switching frequency

## Scaling Features

## Separating Features & Target

```python
X = df.drop('PerformanceRating', axis=1)
y = df['PerformanceRating']
```

## Applying Standard Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

Standardizes features (mean=0, std=1)<br>
Important for models like Logistic Regression

# Final Dataset

```python
X = pd.DataFrame(X_scaled, columns=X.columns)

df_final = pd.concat([X, y.reset_index(drop=True)], axis=1)

df_final
```
