# Uni-Variate Analysis

## Target Variable Analysis (PerformanceRating)

```python
sns.countplot(x='PerformanceRating', data=df)
plt.title("Performance Rating Distribution")
plt.show()
```

Visualizes distribution of target variable

Insight:

Majority employees have rating = 3 (average performers)<br>
Dataset is imbalanced

## Attrition Analysis

```python
sns.countplot(x='Attrition', data=df)
plt.title("Attrition Distribution")
plt.show()
```

Shows employees leaving vs staying

Insight:

Majority employees stay in company<br>
Attrition rate is low (~15%)

## Department-wise Performance

```python
dept_perf = df.groupby('EmpDepartment')['PerformanceRating'].mean().sort_values()

dept_perf.plot(kind='bar')
plt.title("Department vs Performance")
plt.show()
```

Compares average performance across departments

Insight:

All departments consistently perform better<br>
Helps identify weak-performing departments

# Bi-Variate analysis

## Salary Hike vs Performance

```python
sns.boxplot(x='PerformanceRating', y='EmpLastSalaryHikePercent', data=df)
plt.title("Salary Hike vs Performance")
plt.show()
```

Shows relationship between salary hike and performance

Insight:

Higher performers receive higher salary hikes<br>
Strong indicator of performance

## Work-Life Balance vs Performance

```python
sns.boxplot(x='PerformanceRating', y='EmpWorkLifeBalance', data=df)
plt.show()
```

Analyzes effect of work-life balance

Insight:

Better work-life balance → better performance<br>
Poor balance → lower ratings

## Job Satisfaction vs Performance

```python
sns.boxplot(x='PerformanceRating', y='EmpJobSatisfaction', data=df)
plt.show()
```

Relationship between satisfaction and performance

Insight:

High job satisfaction strongly correlates with high performance

## Overtime Impact

```python
sns.countplot(x='OverTime', hue='PerformanceRating', data=df)
plt.show()
```

Shows how overtime affects performance

Insight:<br>

Mixed effect:<br>
Moderate overtime → good performance<br>
Excess overtime → reduced performance<br>

## Experience vs Performance

```python
sns.boxplot(x='PerformanceRating', y='TotalWorkExperienceInYears', data=df)
plt.show()
```

Analyzes experience impact

Insight:<br>

Mid-level experience employees perform best<br>
Too low or too high experience → lower performance<br>

## Distance from Home

```python
sns.boxplot(x='PerformanceRating', y='DistanceFromHome', data=df)
plt.show()
```

Checks if commute affects performance

Insight:<br>

Long distance slightly reduces performance<br>
Not a strong factor

# Multi-Variate analysis

## Correlation Heat map

```python
plt.figure(figsize=(12,8))

numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```

Shows relationships between numerical variables

Insight:<br>

Strong correlations:<br>

Salary hike ↔ Performance<br>
Job level ↔ Income<br>
Helps in feature selection<br>

No extreme multicollinearity → dataset is stable

## Pairplot (Key Features vs Performance)

```python
sns.pairplot(df[['PerformanceRating', 'EmpLastSalaryHikePercent', 
                 'EmpJobSatisfaction', 'EmpWorkLifeBalance', 
                 'TotalWorkExperienceInYears']])
plt.show()
```

Visualizes relationships between multiple variables simultaneously<br>
Shows distribution + relationships

Insight:

Higher salary hike → higher performance cluster<br>
Job satisfaction & work-life balance show positive trend<br>
Mid-range experience shows better performance clustering<br>

## Experience + Job Level + Performance

```python
sns.scatterplot(x='TotalWorkExperienceInYears', y='EmpJobLevel', 
                hue='PerformanceRating', data=df)
plt.title("Experience vs Job Level vs Performance")
plt.show()
```

Multivariate relationship between:<br>
    
Experience<br>
Job level<br>
Performance<br>

Insight:

Mid-experience + mid/high job level → best performers<br>
Very low or very high experience → lower performance

## 3D Relationship (Advanced Insight)

```python
sns.scatterplot(x='EmpLastSalaryHikePercent', 
                y='EmpJobSatisfaction', 
                hue='PerformanceRating', 
                size='EmpWorkLifeBalance',
                data=df)
plt.title("Multi-factor Impact on Performance")
plt.show()
```

Combines multiple features in one visualization:

Salary hike<br>
Satisfaction<br>
Work-life balance<br>

Insight:

High performance occurs when:

Salary hike is high<br>
Job satisfaction is high<br>
Work-life balance is good<br>

**Top 3 Factors Affecting Performance:**
    
- Percent Salary Hike<br>
- Job Satisfaction<br>
- Work-Life Balance<br>

**Key Patterns:**
    
High salary + high satisfaction = high performance<br>
Low satisfaction + high overtime = low performance<br>
Mid-level experience employees perform best

# EDA Summary – Employee Performance Dataset

* The dataset consists of employee demographic, job-related, and satisfaction features with **PerformanceRating as the target variable**.

* Most employees fall under **average performance (Rating = 3)**, indicating a **class imbalance** with fewer high performers.

* **Department-wise analysis** shows variation in performance, highlighting areas with potential improvement.

* **Key factors influencing performance**:

  * Salary Hike %
  * Job Satisfaction
  * Work-Life Balance

* Employees with **higher satisfaction, better work-life balance, and higher salary increments** tend to perform better.

* **Overtime shows mixed impact** — moderate overtime may improve performance, but excessive overtime leads to burnout.

* **Experience plays a role**, with mid-level experienced employees showing better performance compared to very low or very high experience levels.

* Multivariate analysis confirms that **performance is driven by a combination of factors**, not a single variable.
