# DATA_PROCESSING_REPORT

## Employee Performance Analysis – INX Future Inc.

### **1. Executive Summary**
The study was conducted to analyze employee performance at INX Future Inc. (Project Code: 10281) using a dataset of 1,200 employees across 28 distinct features. The primary objective was to process and clean the raw organizational data to prepare it for robust performance modeling. The resulting dataset is a refined collection of categorical and numerical variables covering demographics, job roles, and satisfaction levels. Key insights from the preprocessing phase reveal a high-quality dataset with no missing values or duplicate records, ensuring the integrity of subsequent predictive analysis. Outliers in numerical features were successfully managed to prevent model bias.

### **2. Data Acquisition and Experimental Design**
The experimental design focused on a cross-sectional analysis of the company's workforce.
* **Data Source:** The raw data was acquired from a comprehensive internal record titled `INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls`.
  
* **Sample Size:** The dataset contains 1,200 individual employee observations.
* **Feature Scope:** 28 features were analyzed, including:
    * **Demographics:** Age, Gender, Education Background, Marital Status.
    * **Operational Data:** Department, Job Role, Business Travel Frequency, Distance from Home, Overtime.
    * **Performance & Satisfaction:** Performance Rating (Target), Job Satisfaction, Environment Satisfaction, Relationship Satisfaction, and Work-Life Balance.
    * **Tenure & Compensation:** Total Work Experience, Years at Company, Last Salary Hike Percent, and Hourly Rate.

**There were no Duplicates and Null values present in the datatset**

### **3. Data Preprocessing Methods**
The data underwent a rigorous processing pipeline to ensure suitability for machine learning:


* **Initial Audit:** A comprehensive check for data integrity was performed. The dataset was confirmed to be clean with zero missing values and no duplicate entries.
* **Feature Engineering & Selection:**
  
    * **Feature Dropping:** Irrelevant features that did not contribute to predictive power (such as unique identifiers like `EmpNumber`) were removed from the training set.
    * **Type Casting:** Features were categorized based on their nature. Discrete and ordinal values were maintained as **INT**, while continuous measurable values were stored as **FLOAT** to preserve precision.
      
**Outlier Treatment:** To ensure robust model performance, outliers were identified and treated using the **Interquartile Range (IQR) capping method**. This involved:
    1.  Calculating the IQR (Q3 - Q1).
    2.  Defining bounds (Lower: Q1 - 1.5*IQR; Upper: Q3 + 1.5*IQR).
    3.  Capping extreme values at these bounds to minimize the influence of anomalies.
       
* **Final Export:** The processed and cleaned dataframe was saved to an Excel format for use in the modeling phase of the project.
