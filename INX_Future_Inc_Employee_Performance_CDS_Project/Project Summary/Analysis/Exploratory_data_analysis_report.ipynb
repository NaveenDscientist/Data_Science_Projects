# EXPLORATORY DATA ANALYSIS (EDA) REPORT

## Employee Performance Analysis – INX Future Inc.

---

# 1. Introduction

The objective of this analysis is to explore the employee dataset and identify key factors influencing employee performance. The dataset contains employee demographic details, job-related attributes, and satisfaction metrics, with **PerformanceRating** as the target variable.

---

# 2. Dataset Overview

* **Total Records:** 1200 employees
* **Total Features:** 28 (after cleaning)
* **Target Variable:** PerformanceRating (2 = Low, 3 = Average, 4 = High)
* **Data Quality:** No missing values, duplicates removed, and irrelevant columns dropped

---

# 3. Data Cleaning Summary

The following preprocessing steps were performed:

* Removed duplicate records to ensure data integrity
* Dropped irrelevant columns:

  * EmployeeCount (constant feature)
  * EmployeeNumber (unique identifier)
* Standardized column names for consistency
* Verified no missing values across all features
* Outliers were treated using the IQR capping method

The dataset was found to be clean and suitable for analysis and modeling.

---

# 4. Univariate Analysis

## Performance Rating Distribution

* Majority employees fall under **Rating 3 (Average)** (~73%)
* Low performers and high performers are fewer

**Insight:**
The dataset is **imbalanced**, with most employees performing at an average level.

---

## Attrition Distribution

* Majority employees remain in the company (~85%)
* Attrition rate is relatively low (~15%)

**Insight:**
The company has good retention, but performance issues persist despite low attrition.

---

# 5. Bivariate Analysis

## Salary Hike vs Performance

* Employees with higher salary hikes tend to have higher performance ratings

**Insight:**
Salary increment is a **strong indicator of performance recognition**.

---

## Job Satisfaction vs Performance

* Higher job satisfaction correlates with higher performance

**Insight:**
Satisfied employees are more productive and perform better.

---

## Work-Life Balance vs Performance

* Employees with better work-life balance tend to perform better

**Insight:**
Work-life balance plays a **critical role in productivity**.

---

## Overtime vs Performance

* Moderate overtime improves performance
* Excessive overtime leads to reduced performance

**Insight:**
Overtime has a **non-linear impact** (positive initially, negative later).

---

## Department-wise Performance

* Performance varies across departments

**Insight:**
Some departments require targeted performance improvement strategies.

---

# 6. Multivariate Analysis

## Correlation Analysis

* Positive correlation observed between:

  * Salary hike and performance
  * Job level and income
* Experience-related features show moderate correlation

**Insight:**
No strong multicollinearity → features are suitable for modeling.

---

## Combined Feature Analysis

* High performance occurs when:

  * Salary hike is high
  * Job satisfaction is high
  * Work-life balance is good

**Insight:**
Performance is driven by **combination of multiple factors** rather than a single variable.

---

## Experience vs Performance

* Mid-level experienced employees perform better
* Very low and very high experience groups show lower performance

**Insight:**
Optimal experience level contributes to better performance.

---

# 7. Key Findings

### Top 3 Factors Affecting Performance:

1. **Percent Salary Hike**
2. **Job Satisfaction**
3. **Work-Life Balance**

---

### Other Influencing Factors:

* Overtime (moderate vs excessive)
* Department
* Experience level

---

# 8. Business Insights

* Employee performance is not solely dependent on experience
* Organizational policies (salary, satisfaction, balance) significantly impact performance
* Departments have varying performance levels, indicating uneven resource utilization

---

# 9. Recommendations

Based on the analysis, the following actions are recommended:

* Improve employee satisfaction through engagement programs
* Provide competitive and performance-based salary increments
* Promote better work-life balance policies
* Monitor overtime to avoid burnout
* Focus on mid-career employee development
* Identify and support underperforming departments

---

# 10. Conclusion

The EDA reveals that employee performance is influenced by a combination of compensation, satisfaction, and work-life balance factors. Organizations can significantly improve performance by focusing on these key areas rather than relying solely on experience or tenure.

---

# Final Statement

> The exploratory data analysis successfully identified the critical drivers of employee performance and provided actionable insights that can help the organization improve productivity and employee satisfaction.
