### 1. Study Overview
The study involved a comparative performance analysis of four distinct classification algorithms: **Logistic Regression, Decision Tree, Random Forest, and Gradient Boosting**. The goal was to determine the most accurate model for predicting employee performance ratings (Levels 2, 3, and 4) based on historical organizational data.

### 2. Experimental Design & Methods
The experimental workflow was designed to ensure objectivity and reproducibility:

* **Data Acquisition & Split:** The dataset was partitioned into training and testing sets. The test data ($X\_test$ and $y\_test$) was stored in `.xlsx` format to ensure data integrity during the evaluation phase.
* **Pre-processing:** While the prediction notebook focuses on evaluation, the underlying models were trained using pre-processed features. This included converting categorical variables into numerical formats and handling potential scaling requirements for distance-based algorithms like Logistic Regression.
* **Model Loading:** Serialized model objects (`.pkl`) were loaded using the `joblib` library to maintain the exact state of the trained parameters.
* **Evaluation Metrics:** The design utilized multiple metrics to provide a 360-degree view of model health:
    * **Accuracy:** Overall correctness.
    * **F1-Score:** Balancing precision and recall, especially important for the minority classes (Level 2 and 4).
    * **Confusion Matrices:** Visualizing misclassifications between adjacent performance levels.

---

### 3. Resulting Data & Model Performance
The evaluation yielded the following accuracy results on the test data:

| Model | Accuracy Score |
| :--- | :--- |
| **Gradient Boosting** | **94.17%** |
| **Decision Tree** | **93.75%** |
| **Random Forest** | **89.58%** |
| **Logistic Regression** | **82.08%** |


### 4. Key Insights
* **Ensemble Superiority:** **Gradient Boosting** emerged as the champion model. It demonstrated high precision ($1.00$) for high-performing employees (Level 4), making it highly reliable for identifying top talent without false positives.
* **Non-Linearity:** The significant performance gap between Logistic Regression ($82\%$) and Gradient Boosting ($94\%$) suggests that the factors influencing employee performance are complex and non-linear.
* **Class Imbalance Impact:** Models performed best on "Level 3" (the majority class), with some drop-off in recall for "Level 2" and "Level 4". This indicates that while the models are highly accurate overall, specific attention is needed when predicting outliers in performance.
* **Decision Stability:** The Decision Tree's high performance ($93.75\%$) suggests that a few key features (e.g., salary hike, environment satisfaction) carry significant weight in determining an employee's performance tier.
