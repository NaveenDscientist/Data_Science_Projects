# TRAIN_MODEL REPORT

## Employee Performance Analysis – INX Future Inc.

### 1. Summary
This study was performed to analyze and predict employee performance ratings at INX Future Inc. using a dataset of 1,200 employee records with 57 features. The project implemented a machine learning pipeline that included feature selection and class balancing to address data complexities. Four primary models—Logistic Regression, Decision Tree, Random Forest, and Gradient Boosting—were trained and evaluated.

**Key Resulting Data and Insights:**
* **Top Performance**: Gradient Boosting achieved the highest cross-validation score of **0.934**, making it the most effective model for this analysis.
* **Model Comparison**: Decision Tree (0.887) and Random Forest (0.881) also showed strong results, while Logistic Regression (0.811) served as a solid baseline.
* **Data Strategy**: The combination of feature selection and the Synthetic Minority Over-sampling Technique (SMOTE) was critical in handling class imbalance and improving overall predictive accuracy.


### 2. Methods

#### Data Acquisition and Initial Setup
* **Data Source**: The dataset was acquired from a processed Excel file containing 1,200 rows and 57 columns.
* **Target Variable**: The model aims to predict `PerformanceRating`.
* **Initial Processing**: Features and target variables were separated, followed by a data split into training (80%) and testing (20%) sets. A `stratify=y` parameter was used to maintain the original class distribution across sets.

#### Computational Preprocessing
* **Feature Selection**: To focus on the most impactful data, a `RandomForestClassifier` was first used to calculate feature importance. `SelectFromModel` with a 'median' threshold was then applied to transform the training and test sets, retaining only the most significant predictors.
* **Normalization/Standardization**: While the final notebook shows pre-scaled data (indicated by negative and decimal values in features like `Age` and `DistanceFromHome`), the pipeline focuses on structural refinements.
* **Class Balancing (SMOTE)**: To address the imbalance in performance rating categories, **SMOTE** (Synthetic Minority Over-sampling Technique) was applied to the training data to generate synthetic samples for minority classes.

#### Experimental Design
* **Cross-Validation**: A **Stratified 3-Fold Cross-Validation** (using `StratifiedKFold`) was implemented with shuffling to ensure the models were robust and generalize well to unseen data.
* **Hyperparameter Tuning**: `GridSearchCV` was employed across all models to find the optimal parameters (e.g., `learning_rate` and `n_estimators` for Gradient Boosting).
* **Model Training**: The training process compared four diverse algorithms:
    * **Logistic Regression**: Fast baseline model.
    * **Decision Tree**: Included for interpretability.
    * **Random Forest**: Chosen for stability and strength.
    * **Gradient Boosting**: Selected for high-performance requirements.
